## g. Determine quartiles and remove outliers

```python
for col in data1.columns:
    q1=data1[col].quantile(0.25)
    q3=data1[col].quantile(0.75)
    iqr=q3-q1
    low=q1-(1.5*iqr)
    high=q3+(1.5*iqr)
    data1 = data1.loc[(data1[col] > low) & (data1[col] < high)]
    
print(data1)
```

## e. Employ Bokeh for interactive data visualization.

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
from bokeh.io import output_notebook


col1 = 'Pregnancies'
col2 = 'Glucose'

output_notebook()

p = figure(title=f'Interactive Scatter Plot of {col1} vs {col2}', 
           x_axis_label=col1, y_axis_label=col2,
           tools='pan,box_zoom,reset,hover,save')

scatter = p.circle(x=df[col1], y=df[col2], size=8, color='red', alpha=0.7)

hover = HoverTool()
hover.tooltips = [(f'{col1}', f'@{col1}'), (f'{col2}', f'@{col2}')]
p.add_tools(hover)

show(p)
```

## f.Utilize Bokeh's GUI features, including buttons, sliders, checkboxes, and radio buttons.

```python
from bokeh.io import curdoc
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, Button, Slider, CheckboxGroup, RadioGroup
from bokeh.layouts import column, row


source = ColumnDataSource(data1)

p = figure(title="Interactive Scatter Plot", x_axis_label='BLOCKS', y_axis_label='VILLAGES')
scatter = p.circle(x='Pregnancies', y='Glucose', source=source, size=8, color='blue', alpha=0.7)

hover = HoverTool()
hover.tooltips = [("Pregnancies", "@Pregnancies"), ("Glucose", "@Glucose")]
p.add_tools(hover)

button = Button(label="Randomize Data", button_type="success")
slider = Slider(title="Size", value=8, start=1, end=20, step=1)
checkbox_group = CheckboxGroup(labels=["Show Scatter Plot"], active=[0])
radio_group = RadioGroup(labels=["Blue", "Red"], active=0)

def update_plot():
    scatter.glyph.size = slider.value

button.on_click(update_plot)
slider.on_change('value', lambda attr, old, new: update_plot())

layout = column(
    row(button, slider),
    row(checkbox_group, radio_group),
    p
)

curdoc().add_root(layout)
show(layout, notebook_handle=True)
```

## g. Employ Plotly to create a dropdown menu.

```python
import plotly.graph_objects as px 

np.random.seed(42) 
  
random_x = np.random.randint(1, 101, 100) 
random_y = np.random.randint(1, 101, 100) 
  
plot = px.Figure(data=[px.Scatter( 
    x=data1['Pregnancies'], 
    y=data1['Glucose'], 
    mode='markers',) 
]) 
  
plot.update_layout( 
    updatemenus=[ 
        dict( 
            buttons=list([ 
                dict( 
                    args=["type", "scatter"], 
                    label="Scatter Plot", 
                    method="restyle"
                ), 
                dict( 
                    args=["type", "bar"], 
                    label="Bar Chart", 
                    method="restyle"
                ) 
            ]), 
            direction="down", 
        ), 
    ] 
) 
  
plot.show() 
```

## h. Develop custom action buttons using Plotly.

```python
import plotly.graph_objects as px 
import numpy 

x = data1['BMI']

plot = px.Figure(data=[px.Bar( 
    name='Pregnancies', 
    x=x, 
    y=data1['Pregnancies'] 
), 
    px.Bar( 
    name='Glucose', 
    x=x, 
    y=data1['Glucose'] 
) 
]) 


plot.update_layout( 
    updatemenus=[ 
        dict( 
            type="buttons", 
            direction="left", 
            buttons=list([ 
                dict(label="Both", 
                    method="update", 
                    args=[{"visible": [True, True]}, 
                        {"title": "Both"}]), 
                dict(label="Pregnancies", 
                    method="update", 
                    args=[{"visible": [True, False]}, 
                          {"title": "Data 1", 
                          }]), 
                dict(label="Glucose", 
                     method="update", 
                     args=[{"visible": [False, True]}, 
                           {"title": "Data 2", 
                           }]), 
            ]), 
        ) 
    ]) 

plot.show() 
```
