## Linear Regression - using BuildIn Functions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Using dataset

```python
df = pd.read_csv("dataset.csv")
df
```

```python
arr1 = np.array(df)
arr1
```

```python
x1 = arr1[:, 0].reshape((-1,1))
y1 = arr1[:, 1].reshape((-1,1))
```

```python
reg1 = LinearRegression()
reg1.fit(x1, y1)
```

```python
#get coeff
print(reg1.coef_)
print(reg1.intercept_)
```

```python
#predict val
y_pred1 = reg1.predict(x1)
print(y_pred1)
```

```python
#check for week
print(reg1.predict([[7]]))
print(reg1.predict([[12]]))
```

```python
plt.scatter(x1, y1)
plt.plot(x1, y_pred1)

plt.title("Linear Regression")
plt.ylabel("y-axis")
plt.xlabel("x-axis")

for i, j in zip(x1, y1):
   plt.text(i, j+0.5, '({}, {})'.format(i, j))

plt.show()
```

## Using random - int

```python
arr2 = np.random.randint(low = 1, high = 10, size =(5, 1))
# making a new y column and multiplying it with the previously build x column 
# (in arr2) to get smaller values for y for small values of x and vice versa 
y = np.multiply(np.random.randint(low = 1, high = 3, size =(5, 1)),arr2)

arr2 = np.append(arr2, y, axis=1)
arr2
```

```python
x2 = arr2[:, 0].reshape((-1,1))
y2 = arr2[:, 1].reshape((-1,1))
```

```python
reg2 = LinearRegression()
reg2.fit(x2, y2)
```

```python
#get coeff
print(reg2.coef_)
print(reg2.intercept_)
```

```python
#predict val
y_pred2 = reg2.predict(x2)
print(y_pred2)
```

```python
#check for week
print(reg2.predict([[7]]))
print(reg2.predict([[12]]))
```

```python
plt.scatter(x2, y2)
plt.plot(x2, y_pred2)

plt.title("Linear Regression")
plt.ylabel("y-axis")
plt.xlabel("x-axis")

for i, j in zip(x2, y2):
   plt.text(i, j+0.5, '({}, {})'.format(i, j))

plt.show()
```
